```python
import os
from PIL import Image
```

```python
door_round_imgs = './Door_imgs/round/'
img_list = os.listdir(door_round_imgs)
if not os.path.isdir(door_round_imgs+'resize/'):
    os.mkdir(door_round_imgs+'resize/')
for f_name in img_list:
    print(f_name)
    if not (f_name=='.DS_Store' or f_name=='resize'):
        image = Image.open(door_round_imgs+f_name)
#         resize_image = image.resize((256,256))
#         resize_image.save(door_round_imgs+'resize/'+f_name+'_resize.jpg')
        desired_size = 256
        # im_pth = "/home/jdhao/test.jpg"

        # im = Image.open(im_pth)
        im = image
        old_size = im.size  # old_size[0] is in (width, height) format

        ratio = float(desired_size)/max(old_size)
        new_size = tuple([int(x*ratio) for x in old_size])
        # use thumbnail() or resize() method to resize the input image

        # thumbnail is a in-place operation

        # im.thumbnail(new_size, Image.ANTIALIAS)

        im = im.resize(new_size, Image.ANTIALIAS)
        # create a new image and paste the resized on it

        new_im = Image.new("RGB", (desired_size, desired_size))
        new_im.paste(im, ((desired_size-new_size[0])//2,
                            (desired_size-new_size[1])//2))
        for i in range(4):
            new_im = new_im.rotate(90*i)
            new_im.save(door_round_imgs+'resize/'+f_name+'_resize_rot'+str(90*i)+'.jpg')
```
